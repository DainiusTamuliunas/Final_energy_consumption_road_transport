```python
import pandas as pd

file_path = '/content/drive/MyDrive/Master Paper/Data/'
```

```python
import json
# Parse the JSON data
# Load car_brands.json
with open(file_path  + 'Json Cars Motorbikes Brand and Model/' + 'car_brands.json', 'r') as file:
    car_brands = json.load(file)

# Load car_models.json
with open(file_path  + 'Json Cars Motorbikes Brand and Model/' + 'car_models.json', 'r') as file:
    car_models = json.load(file)

# Convert car brands and car models data into dataframes
car_brands_df = pd.DataFrame(car_brands["data"])
car_models_df = pd.DataFrame(car_models["data"])

# Merge on the brand_id and id to get brand names alongside models
merged_df = car_models_df.merge(car_brands_df, left_on='brand_id', right_on='id')

# Select only 'name_x' (model) and 'name_y' (brand) columns
final_df = merged_df[['name_y', 'name_x']].rename(columns={'name_y': 'brand', 'name_x': 'model'})

# Save the final dataframe to CSV
final_df.to_csv(file_path + 'unique_car_makers_models_list.csv', index=False)
```

```python
# Load moto_brands.json
with open(file_path  + 'Json Cars Motorbikes Brand and Model/' + 'moto_brands.json', 'r') as file:
    moto_brands = json.load(file)

# Load moto_models.json
with open(file_path  + 'Json Cars Motorbikes Brand and Model/' + 'moto_models.json', 'r') as file:
    moto_models = json.load(file)

# Convert moto brands and moto models data into dataframes
moto_brands_df = pd.DataFrame(moto_brands["data"])
moto_models_df = pd.DataFrame(moto_models["data"])

# Merge on the brand_id and id to get brand names alongside models
merged_df = moto_models_df.merge(moto_brands_df, left_on='brand_id', right_on='id')

# Select only 'name_x' (model) and 'name_y' (brand) columns
final_df = merged_df[['name_y', 'name_x']].rename(columns={'name_y': 'brand', 'name_x': 'model'})

# Save the final dataframe to CSV
final_df.to_csv(file_path + 'unique_moto_makers_models_list.csv', index=False)

csv_file_path = file_path + 'unique_moto_makers_models_list.csv'
print(f'CSV file saved at: {csv_file_path}')
```
